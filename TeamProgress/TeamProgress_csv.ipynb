```python
#Подключаем необходимые библиотеки
import pandas as pd #pandas
import numpy as np #numpy

#Задаем переменные
data_file = 'data\\TeamProgress_Sheet1_2020-05-01.csv' #Расположение файла c данными
myteam_file = 'data\\team06.csv' #Расположение файла с составом команды
course = 'PYRT-3+22APR2020' #Название курса

#Читаем файлы с данными
data = pd.read_csv(data_file, delimiter=',')
myteam = pd.read_csv(myteam_file, delimiter=';', index_col='email')

#Переводим email'ы в нижний регистр для единообразного вида индексов
data['email'] = data['email'].str.lower() # для data
myteam.index.map(str.lower) # для myteam, email - индексное поле

#Склеиваем датафреймы data и myteam
data_team = pd.merge(data, myteam, how='inner', left_on='email', right_index=True)

#Оставляем только данные участников команды, относящиеся к курсу course
data1 = data_team[data_team['course'] == course]

#Создаем сводную таблицу
pd.pivot_table(data1, values = 'ix_homework', index = ['name'], columns = ['homework_sort'], aggfunc = np.sum, margins = True, margins_name = 'Всего')
```
